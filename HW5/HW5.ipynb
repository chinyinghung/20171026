# Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料


```python
import requests
```

```python
url = "http://finance.google.com/finance/historical"
```

```python
params = {}
params['q'] = 'TSLA'
params['startdate'] = '2016-05-01'
params['output'] = 'csv'
```

```python
stockprice = requests.get(url,params = params)
```

```python
print (stockprice.text)
```
